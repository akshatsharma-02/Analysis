```python
import pandas as pd
import matplotlib.pyplot as plt

df= pd.read_csv('D:\\Projects\\Python Pandas\\Amazon_Electronics.csv')
df.head()

rating_group_by_category= df.groupby(['category','rating']).size()
print(rating_group_by_category,"\n")

products= df.groupby('category').size()
print(products,"\n")

female_products_count= df['model_attr'].value_counts()['Female']
print("No.of female products are:",female_products_count)
male_products_count= df['model_attr'].value_counts()['Male']
print("No.of male products are:",male_products_count)
both_products_count= df['model_attr'].value_counts()['Female&Male']
print("No.of female and male products are:",both_products_count)

average_rating_per_category = df.groupby('category')['rating'].mean()
plt.figure(figsize=(10, 6))
average_rating_per_category.plot(kind='bar')
plt.title("Average Ratings by Category")
plt.xlabel("Category")
plt.ylabel("Average Rating")
plt.show()

model_attr_counts= df['model_attr'].value_counts()
plt.figure(figsize=(8, 8))
model_attr_counts.plot(kind='pie',autopct='%1.1f%%')
plt.title("Distribution of Products by Model Attribute")
plt.ylabel("")
plt.show()
```
